**Problem 2**

The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information; (Provide an appropriate visual representation of your answers)



# **Importing neccesary libraries**

```python
#import the required function
from scipy.stats import norm
# import the important packages
import pandas as pd  # library used for data manipulation and analysis

import numpy as np  # library used for working with arrays

import matplotlib.pyplot as plt  # library for visualizations

import seaborn as sns  # library for visualizations

%matplotlib inline

import scipy.stats as stats # this library contains a large number of probability distributions as well as a growing library of statistical function
```

Given:
       
       Population Mean(µ) = 5 kg/cm^2.
       Population standard Deviation(σ) = 1.5 kg/cm^2.
       The given data is normally distributed.

```python
mu=5
sigma=1.5
#plotting between 0 to 1 with
#0.1 steps.
x_axis=np.arange(0, 10, 0.1)
```

```python
#Calculating mean and standard deviation
mean=mu
sd=sigma
```

**2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?**

**Solution**

```python
#To calculate this, we will calculate the cumulative probability for less than 3.17 kg/cm^2
norm = stats.norm.cdf(3.17,loc=5,scale=1.5)
print((round(norm, 4)*100), "%")
```

**Let's Plot the normal distribution**

```python
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.axvline(x=3.17, c="r")
plt.xlabel("Breaking strength of gunny bags kg/cm^2")
plt.show()
```

**2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?**

```python
#To calculate this, we will calculate the cumulative probability for less than 3.6 kg/cm^2 and then will subtract from 1.
norm = 1-stats.norm.cdf(3.6,loc=5,scale=1.5)
print((round(norm, 4)*100), "%")
```

**Let's Plot the normal distribution**

```python
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.axvline(x=3.6, c="r")
plt.xlabel("Breaking strength of gunny bags kg/cm^2")
plt.show()
```

**2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?**

```python
norm = stats.norm.cdf(5.5,loc=5,scale=1.5) - stats.norm.cdf(5,loc=5,scale=1.5)
print((round(norm, 4)*100), "%")
```

**Let's Plot the normal distribution**

```python
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.axvline(x=5, c="r")
plt.axvline(x=5.5, c="g")
plt.xlabel("Breaking strength of gunny bags kg/cm^2")
plt.show()
```

**2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?**

```python
norm = 1-(stats.norm.cdf(7.5,loc=5,scale=1.5) - stats.norm.cdf(3,loc=5,scale=1.5))
print((round(norm, 4)*100), "%")
```

**Let's Plot the normal distribution**

```python
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.axvline(x=3, c="r")
plt.axvline(x=7.5, c="g")
plt.xlabel("Breaking strength of gunny bags kg/cm^2")
plt.show()
```

**3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?**


To test whether Zingaro is justified in thinking that the unpolished stones may not be suitable for printing, we need to perform a one-sample t-test for the unpolished stones.






# Let's write the null and alternative hypothesis
The null hypothesis for this test is that the mean hardness of the unpolished stones is greater than or equal to 150, while the alternative hypothesis is that the mean hardness is less than 150.

We will test the null hypothesis

>$H_0:\mu >= 150$

against the alternate hypothesis

>$H_a:\mu < 150$

Level of significance(α)=0.05

We can perform 1 sample z-test as this is a one-sided sample with **sample size 75>30**

μ=150 and n=75

Also, X̄1 unpolished= 134.11, σ1 unpolished = 33.04, X̄2 polished= 147.79, σ2 polished = 15.59.




### Let's test whether the Z-test assumptions are satisfied or not

* Continuous data - Brinell's hardness index is measured on a continuous scale.
* Normally distributed Sample sizes > 30 - Since the sample sizes are greater than 30, Central Limit Theorem states that the distribution of sample means will be normal.
* Independent Sample sizes - As we are taking samples for two different type of stones, the two samples are from two independent populations.
* Known population standard deviation $\sigma_1$ and $\sigma_2$ - Yes, we know the population standard deviations of both the populations.
* Random sampling from the population - Yes, we are informed that the collected sample is a simple random sample.

We can use 1 sample Z-test for this problem.


### Let's find the p-value

We are going to use the below formula to calculate the z score of the test using user defined function.


```python
# setting the values of population mean and standard deviation to 150 and 33.04 respectively for unpolished stones
mu, sigma = 150, 33.04

# set the value sample mean to 134.11
x1_bar = 134.11

# calculate the test statistic
test_stat = (134.11 - 150) / (33.04/np.sqrt(75))
test_stat
```

```python
# calculate the one-tailed p-value for unpolished stones
pval = 1 - norm.cdf(abs(test_stat))
print('The P-value is', pval)
```

```python
# setting the values of population mean and standard deviation to 150 and 15.59 respectively for treated and polished stones
mu, sigma = 150, 15.59

# set the value sample mean to 147.79
x2_bar = 147.79

# calculate the test statistic
test_stat = (147.79 - 150) / (15.59/np.sqrt(75))
test_stat
```

```python
# calculate the one-tailed p-value for treated and polished stones
pval = 1 - norm.cdf(abs(test_stat))
print('The P-value is', pval)
```

### Insights
p-value for unpolished~0 < 0.05 (i.e. α) and hence Null Hypothesis is rejected in
that case. So, unpolished stones do not have a Brinell's hardness index of at
least 150. While, in case of treated and polished p-value is 0.11 > 0.05 (i.e. α) and
hence Null Hypothesis cannot be rejected in that case. So, treated and polished
stones have a Brinell's hardness index of at least 150.


**CONCLUSION 1:**

Hence, Zingaro is right in thinking that unpolished stones are not

**3.2 Is the mean hardness of the polished and unpolished stones the same?**


# Let's write the null and alternative hypothesis

Null Hypothesis (H0) : mean polished = mean unpolished

Alternate Hypothesis (Ha): mean polished ≠
mean unpolished

We will test the null hypothesis

 H0 :μ1=μ2

against the alternate hypothesis

Ha :μ1≠μ2

Level of significance(α)=0.05

We can perform Two-tailed t-test, Also, X̄1 unpolished= 134.11, σ1 unpolished = 33.04, X̄2 polished= 147.79, σ2 polished = 15.59, n=75

### Let's find the p-value

```python
# Two-sample t-test for polished and unpolished stones
#μ=150 and n1=75, n2=75, X̄1 unpolished= 134.11, σ1 unpolished=33.04, X̄2 polished=147.79, σ2 polished =15.59
sp = ((75-1)*15.59**2 + (75-1)*33.04**2) / (75+75-2)
sp
```

```python
test_stat = (147.79-134.11) / (sp * ((1/75) + (1/75)))**0.5
test_stat
```

```python
p_value = stats.t.sf(abs(test_stat), 75+75-2)
p_value
```

### Insights
The p-value of the two-tailed test is 0.00073128 and is significantly less than 0.05 (i.e.
α), so alternate hypothesis prevails.

**CONCLUSION 2:**

hardness of dental implants among different dentists.

# **Importing neccesary libraries**

```python
# import the important packages
import pandas as pd  # library used for data manipulation and analysis
import numpy as np
import scipy.stats as stats # this library contains a large number of probability distributions as well as a growing library of statistical function
import statsmodels.api as smf
from statsmodels.stats.anova import anova_lm
from statsmodels.formula.api import ols # For n-way ANOVA
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.sandbox.stats.multicomp import GroupsStats
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro
```

```python
#Loading the datasets
data ="/content/sample_data/Dental+Hardness+data.csv"
df = pd.read_csv(data)
#separate the data by alloy type
# Assuming data is stored in dataframes df_alloy1 and df_alloy2 for each type of alloy
df_alloy1 = data[data[1]==1]
df_alloy2 = data[data[2]==2]
```

#Assumptions of the Hypothesis Test



*   **Independence:** The hardness measurements for each dentist are
independent of each other.

*   **Normality:** The hardness measurements for each dentist follow a normal distribution.



*   **Homogeneity of Variance:** The variance of hardness measurements is equal across different dentists.






#Conducting the Hypothesis Test and Computing the P-Value

**Hardness by Dentists**:

The code uses ANOVA to determine if there's a statistically significant difference in hardness based on the dentist.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols

# ANOVA test for hardness by dentists
model_dentist = ols('Response ~ Dentist', data=df).fit()
anova_table_dentist = sm.stats.anova_lm(model_dentist, typ=2)
print("ANOVA Table - Dentist:\n", anova_table_dentist)
```

**Conclusions from the Test Results**

If the p-value is less than the chosen significance level (e.g., 0.05), we reject the null hypothesis and conclude that there is a significant difference in the mean hardness of dental implants among different dentists.
If the p-value is greater than the significance level, we fail to reject the null hypothesis.

**Identifying Pairs with Different Implant Hardness**

If the null hypothesis is rejected, post-hoc tests (e.g., Tukey's HSD test) can be conducted to identify which specific pairs of dentists show significant differences in implant hardness.

By following these steps, we can analyze the variation in implant hardness based on dentists and determine if there are significant differences among them.

 **4.3 What is the interaction effect between the dentist and method on the hardness of dental implants for each type of alloy?**

**Interaction Effect:**

This analysis checks if there is an interaction effect between dentist, method, and alloy on hardness.

```python
# Create an interaction plot for Dentist and Method
model_interaction = ols('Response ~ Dentist * Method * Alloy', data=df).fit()
anova_table_interaction = sm.stats.anova_lm(model_interaction, typ=2)
print("\nANOVA Table - Interaction (Dentist * Method * Alloy):\n", anova_table_interaction)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Visualizing the interaction effect
sns.pointplot(data=df, x='Method', y='Response', hue='Dentist', dodge=True, markers=['o', 's', 'D'], linestyles=['-', '--', '-.'])
plt.title('Interaction Effect of Dentist and Method on Response')
plt.show()
```

 **Null hypothesis(H0):** There is no interaction effect between dentists and methods on hardness

**Hardness by Dentists and Methods Together:**

This ANOVA checks if there is a combined effect of dentists and methods on hardness.
